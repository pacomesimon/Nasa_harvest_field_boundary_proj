```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
metrics_table=pd.read_csv("Unet_autoenc_batchsize_4.csv")
def plot_metric(metric,label):
  plt.plot(metrics_table.index, metrics_table[metric],label=label)
  plt.xlabel("Epochs")
  plt.ylabel("Value")
plt.figure()
plot_metric("val_loss","Loss")
plot_metric("val_f1","F1 score")
plt.title("Validation metrics")
plt.legend(title='Metrics:')
plt.show()
# val_loss 	val_precision 	val_recall 	val_f1 	val_accuracy
```
